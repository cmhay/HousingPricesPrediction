Exploratory Data Analysis for housing prices prediction dataset.

```python
import numpy as np
import pandas as pd
# import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display

train = pd.read_csv('./data/train.csv')
# test_data = np.array(test)
# pd.DataFrame(test)

train.head()
```

```python
print('number of rows in data: ' + str(train.shape[0]))
print('number of cols in data: ' + str(train.shape[1]))
```

```python
# get number of NA values for each column
na_col = train.isnull().sum()

# show only columns with at least 1 NA value
na_col = na_col[na_col.values > 0]
print('number of columns with 1+ NA values: len(na_col): ' + str(len(na_col)))

# show the count for each 
display(pd.DataFrame(na_col.sort_values(ascending=False)))

na_col.sort_values(ascending=True).plot(kind='barh', figsize=(8,10))
plt.show()
```
